```python
import numpy as np
from numpy.linalg import norm
from symmetrize import pointops as po, sym
import matplotlib.pyplot as plt
```

### 1. Generate point pattern

```python
sq = np.sqrt(2)
cent = np.array([0., 0.]) + 0.4

# Generate point pattern using a fixed center, a starting point (vertex), and a set of angles of rotation
qs = sym.rotVertexGenerator(cent, fixedvertex=np.array([sq, -sq]), cvd=0.5*sq, arot=[90, 180, 270])

# Order the points
qs_ord = po.pointset_order(qs, direction='ccw')
```

### 2. Calculate the continuous symmetry measure (CSM)
The expected value for a completely symmetric pattern is ~ 0 (up to numerical errors)

```python
po.csm(cent, qs_ord, rotsym=4, type='rotation')
```

### 3. Visualize pattern 

```python
qs_aug = np.concatenate((qs_ord, qs_ord[0:1,:]), axis=0)
plt.figure(figsize=(5,5))
plt.scatter(qs[:,1], qs[:,0], s=50)
plt.axvline(x=0, linestyle='--')
plt.axhline(y=0, linestyle='--')
for i in range(4):
    plt.text(qs_ord[i,1], qs_ord[i,0], str(i+1), size=15)
plt.plot(qs_aug[:,1], qs_aug[:,0], c='k', linestyle='-', linewidth=0.4)
# plt.savefig('../resources/figures/Square_pattern.png', bbox_inches='tight', dpi=300)
```
